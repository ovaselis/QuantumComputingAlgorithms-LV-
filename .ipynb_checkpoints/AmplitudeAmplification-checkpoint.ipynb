# AmplitЕ«du pastiprinДҒЕЎana (Amplitude Amplification) 

AmplitЕ«du pastiprinДҒЕЎana ir vispДҒrinДҒjums Grovera meklД“ЕЎanai. TДҒ Дјauj palielinДҒt вҖңlabДҒвҖқ (atzД«mД“tДҒ) rezultДҒta varbЕ«tД«bu, ja jau ir pieejams kvantu algoritms $A$, kas risinДҒjumu atrod ar kДҒdu varbЕ«tД«bu $p$.

Ja algoritms $A$ reizД“m dod pareizo atbildi, tad ar Д«paЕЎu refleksiju (spoguДјoЕЎanas) operatoru palД«dzД«bu mД“s varam вҖңpagrieztвҖқ kvantu stДҒvokli tДҒ, lai varbЕ«tД«ba iegЕ«t pareizo atbildi kДјЕ«tu lielДҒka.


## Pamatideja

Mums ir unitДҒrs operators $A$, kas no sДҒkuma stДҒvokДјa $|0\rangle$ sagatavo stДҒvokli $|\psi\rangle = A|0\rangle.$

Е ajДҒ stДҒvoklД« daДјa amplitЕ«du ir uz "marД·Д“tajiem" stДҒvokДјiem un daДјa uz "nemarД·Д“tajiem" stДҒvokДјiem.

- $|\psi_g\rangle$ - normalizД“ta "marД·Д“to" stДҒvokДјu apakЕЎtelpa,
- $|\psi_b\rangle$ - normalizД“ta "nemarД·Д“to" stДҒvokДјu apakЕЎtelpa.

Tad

$$
|\psi\rangle = \sqrt{p}\,|\psi_g\rangle + \sqrt{1-p}\,|\psi_b\rangle,
$$

kur $p$ ir varbЕ«tД«ba, ka mД“rД«jumДҒ iegЕ«stam kДҒdu marД·Д“to stДҒvokli.

IevieЕЎam leЕҶД·i $\theta$, tДҒ lai

$$
\sin^2\theta = p,
\qquad 0<\theta\le \frac{\pi}{2}.
$$

Tad varam rakstД«t:

$$
|\psi\rangle = \sin\theta\,|\psi_g\rangle + \cos\theta\,|\psi_b\rangle.
$$

Е Д« ir 2-dimensionДҒla ДЈeometrija plaknД“, ko veido $|\psi_g\rangle$ un $|\psi_b\rangle$.

## Refleksijas

Amplitude amplification balstДҒs uz diviem refleksijas operatoriem.

### Oracle refleksija

DefinД“jam fДҒzes orДҒkulu:

$$
S_f |x\rangle = (-1)^{f(x)} |x\rangle.
$$

Tas ir atbildД«gs par to, lai
- "marД·Д“tajiem" stДҒvokДјiem fДҒze kДјЕ«st $-1$,
-  pДҒrД“jiem stДҒvokДјiem nekas nemainДҒs.

### Refleksija ap sДҒkuma stДҒvokli

DefinД“jam refleksiju ap $|0\rangle$:

$$
S_0 = I - 2|0\rangle\langle 0|.
$$

Refleksiju ap $|\psi\rangle = A|0\rangle$ var realizД“t kДҒ:

$$
S_\psi = A S_0 A^\dagger.
$$

DefinД“jam operatoru:

$$
Q = -A S_0 A^\dagger S_f.
$$

Tas ir divu refleksiju salikums.  
Divu refleksiju kompozД«cija ДЈeometriski atbilst rotДҒcijai.

## RotДҒcijas interpretДҒcija

SДҒkotnД“jais stДҒvoklis:

$$
|\psi\rangle
=
\sin(\theta)\,|\psi_{\text{good}}\rangle
+
\cos(\theta)\,|\psi_{\text{bad}}\rangle.
$$

PД“c $k$ iterДҒcijДҒm iegЕ«stam:

$$
Q^k |\psi\rangle
=
\sin\big((2k+1)\theta\big)\,|\psi_{\text{good}}\rangle
+
\cos\big((2k+1)\theta\big)\,|\psi_{\text{bad}}\rangle.
$$

TДҒtad varbЕ«tД«ba, ka tiks nomД“rД«ts marД·Д“ts stДҒvoklis pД“c $k$ iterДҒcijДҒm ir:

$$
p_k = \sin^2\big((2k+1)\theta\big).
$$

MaksimДҒlДҒ varbЕ«tД«ba tiek sasniegta, ja:

$$
(2k+1)\theta \approx \frac{\pi}{2}.
$$

No tДҒ iegЕ«st aptuveno optimДҒlo iterДҒciju skaitu:

$$
k \approx \left\lfloor \frac{\pi}{4\theta} - \frac{1}{2} \right\rfloor.
$$

Ja $k$ ir pДҒrДҒk liels, tad $(2k+1)\theta$ pДҒrsniedz $\pi/2$ un varbЕ«tД«ba sДҒk kristies.  

## SaistД«ba ar Grovera algoritmu

Grovera meklД“ЕЎana ir Д«paЕЎs amplitЕ«du pastiprinДҒЕЎanas gadД«jums, kur:

- $A$ sagatavo vienmД“rД«gu superpozД«ciju,
- $S_f$ atzД«mД“ meklД“to elementu,
- $Q$ tiek atkДҒrtots optimДҒlu reiЕҫu skaitu.


```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile

# MarД·Д“tais stДҒvoklis ir <101|

qc = QuantumCircuit(3, 3)

# Oracle realizДҒcija  (vispДҒrД«ga fДҒzes apgrieЕЎana konkrД“tam bitstringam)
def phase_flip(qc, qubits, bitstring):
    q0, q1, q2 = qubits
    bits = [int(b) for b in bitstring]

    # PДҒrveido mД“rД·a stДҒvokli uz |111> (kur bit = 0, uzliek X)
    if bits[0] == 0: qc.x(q0)
    if bits[1] == 0: qc.x(q1)
    if bits[2] == 0: qc.x(q2)

    # CCZ realizДҒcija caur H + CCX + H uz pД“dД“jДҒ kubita
    qc.h(q2)
    qc.ccx(q0, q1, q2)
    qc.h(q2)

    # AtceДј iepriekЕЎД“jos X
    if bits[0] == 0: qc.x(q0)
    if bits[1] == 0: qc.x(q1)
    if bits[2] == 0: qc.x(q2)

# sagatavo brД«vi izvД“lД“tu nevienmД“rД«ga superpozД«cija
qc.ry(0.9, 0)
qc.ry(1.7, 1)
qc.ry(1.2, 2)

# IterДҒciju skaits
num_iterations = 2

for i in range(num_iterations):

    # Sf: oracle 
    phase_flip(qc, [0, 1, 2], "101")

    # AвҖ : sagatavoЕЎanas inversais (apgriezta secД«ba un negatД«vi leЕҶД·i)
    qc.ry(-1.2, 2)
    qc.ry(-1.7, 1)
    qc.ry(-0.9, 0)

    # S0: fДҒzes inversija tikai |000>
    
    qc.x([0, 1, 2])

    # CCZ uz |111>
    qc.h(2)
    qc.ccx(0, 1, 2)
    qc.h(2)

    qc.x([0, 1, 2])   # atpakaДј uz oriДЈinДҒlo bДҒzi

    # A: atkal uz priekЕЎu
    qc.ry(0.9, 0)
    qc.ry(1.7, 1)
    qc.ry(1.2, 2)

qc.measure([0, 1, 2], [0, 1, 2])

print(qc.draw())

sim = AerSimulator()
tqc = transpile(qc, sim)
result = sim.run(tqc, shots=2000).result()
counts = result.get_counts()

print("MД“rД«jumu sadalД«jums:", counts)
```
