# Quantum Fourier Transform

**Kvantu Furjē transformācija** (QFT) ir kvantu analogs diskrētajai Furjē transformācijai (DFT). Tā ir unitāra (tātad atgriezeniska) transformācija, kas pārnes kvantu stāvokli no skaitliskās/indeksa bāzes uz “frekvenču” bāzi. Praktiski QFT ir ļoti svarīga algoritmos, kas izmanto interferenci un fāzes informāciju (piem. fāzes novērtēšanā un Šora algoritmā).

### Formulas

Pieņemsim, ka mums ir $N=2^n$ dimensiju kvantu stāvokļa vektors
$
\mathbf{x}=(x_0,x_1,\dots,x_{N-1})^T
$
kas reprezentē stāvokli
$$
|\psi\rangle=\sum_{j=0}^{N-1} x_j |j\rangle.
$$

QFT stāvoklim $|\psi\rangle$ tiek definēta kā
$$
|\phi\rangle=\sum_{k=0}^{N-1} y_k |k\rangle,
$$
kur $y_k$ ir definēts kā iepriekš
$$
y_k=\frac{1}{\sqrt{N}}\sum_{j=0}^{N-1} e^{\frac{2\pi i}{N}jk}\,x_j.
$$

Tātad bāzes stāvoklim $|j\rangle$, kas atbilst vektoram ar $x_j=1$ un visiem pārējiem $x=0$, QFT ir
$$
\mathrm{QFT}_N|j\rangle=\frac{1}{\sqrt{N}}\sum_{k=0}^{N-1} e^{\frac{2\pi i}{N}jk}\,|k\rangle.
$$

$e^{\frac{2\pi i}{N}}$ bieži apzīmē arī ar $\omega,$ un tad to pašu var pierakstīt kā
$$
\mathrm{QFT}_N|j\rangle=\frac{1}{\sqrt{N}}\sum_{k=0}^{N-1}\omega^{jk}\,|k\rangle.
$$



## Piemērs: QFT uz bāzes stāvokļa $|11\rangle$

Ja mums ir $n=2$ kubiti, tātad $N=4$ un $\omega=e^{\frac{2\pi i}{4}}=i.$
Bāzes stāvoklis $|11\rangle$ atbilst skaitlim $j=3$.
$$
\mathrm{QFT}_4|j\rangle=\frac{1}{\sqrt{4}}\sum_{k=0}^{3}\omega^{jk}|k\rangle
=\frac{1}{2}\sum_{k=0}^{3}\omega^{3k}|k\rangle.
$$

Aprēķinām koeficientus katram $k$:
- $k=0$: $\omega^{3\cdot 0}=\omega^0=1$
- $k=1$: $\omega^{3\cdot 1}=\omega^3=i^3=-i$
- $k=2$: $\omega^{3\cdot 2}=\omega^6=i^6=-1$
- $k=3$: $\omega^{3\cdot 3}=\omega^9=i^9=i$

Tātad:
$$
\mathrm{QFT}_4|11\rangle
=\frac{1}{2}\Big(|00\rangle - i|01\rangle - |10\rangle + i|11\rangle\Big).
$$

## Piemērs: QFT uz stāvokļa $\;\dfrac{|01\rangle+|11\rangle}{\sqrt2}$

Pieņemsim $n=2$ kubiti, tātad $N=4$ un $\omega=e^{\frac{2\pi i}{4}}=i.$

Sākuma stāvoklis ir $|\psi\rangle=\frac{|01\rangle+|11\rangle}{\sqrt2}.$

Tā kā QFT ir lineāra
$$
\mathrm{QFT}_4|\psi\rangle
=\frac{1}{\sqrt2}\left(\mathrm{QFT}_4|01\rangle+\mathrm{QFT}_4|11\rangle\right).
$$

$$
\mathrm{QFT}_4|01\rangle
=\frac12\Big(|00\rangle+i|01\rangle-|10\rangle-i|11\rangle\Big),
$$
$$
\mathrm{QFT}_4|11\rangle
=\frac12\Big(|00\rangle-i|01\rangle-|10\rangle+i|11\rangle\Big).
$$

Saskaitām locekļus pa bāzes stāvokļiem:
- pie $|00\rangle$: $\frac12+\frac12=1$
- pie $|01\rangle$: $\frac{i}{2}+\left(-\frac{i}{2}\right)=0$
- pie $|10\rangle$: $-\frac12+(-\frac12)=-1$
- pie $|11\rangle$: $-\frac{i}{2}+\frac{i}{2}=0$

Tāpēc
$$
\mathrm{QFT}_4|01\rangle+\mathrm{QFT}_4|11\rangle
=|00\rangle-|10\rangle.
$$

Gala rezultāts ir
$$
{
\mathrm{QFT}_4\left(\frac{|01\rangle+|11\rangle}{\sqrt2}\right)
=\frac{|00\rangle-|10\rangle}{\sqrt2}.
}
$$


## Realizācija Python
Ir vairāki veidi, kā var realizēt QPE iekš Python, bet katrs no tiem var tikt pielietots, dažādām situācijām. Šeit ir vienkārš piemērs izmantojot NumPy, lai izveidotu QFT funkciju, kas pēc formulas aprēķina tos pašus piemērus, ko iepriekš.

```python
import numpy as np
# Importējam NumPy bibliotēku, lai ērti strādātu ar masīviem un kompleksiem skaitļiem.

def qft(x):
    x = np.array(x, dtype=complex)
    # Pārvēršam ievadi x par NumPy masīvu ar kompleksu tipu.
    N = len(x)
    y = np.zeros(N, dtype=complex)
    # Izveidojam tukšu rezultāta masīvu y ar N elementiem.
    # Sākumā visi ir 0+0j, jeb kompleksās nulles
    for k in range(N):
        # cilpa iet cauri visiem izejas indeksiem k = 0..N-1.
        # aprēķināsim katru y_k atsevišķi.
        s = 0
        # s ir summa
        for j in range(N):
            # Iekšējā cilpa iet cauri visiem ievades indeksiem j = 0..N-1.
            s += np.exp(2j * np.pi * j * k / N) * x[j]
        y[k] = s / np.sqrt(N)
    return y
    #Atgriežam visu rezultāta vektoru y = [y_0, y_1, ..., y_{N-1}]

# Piemērs |11>
x_11 = [0, 0, 0, 1]
y_11 = qft(x_11)
print("Ievade |11> amplitūdas:", x_11)
print("QFT(|11>) amplitūdas:", np.round(y_11, 3)) 


#Piemērs (|01> + |11>)/sqrt(2)
x_super = [0, 1/np.sqrt(2), 0, 1/np.sqrt(2)]
y_super = qft(x_super)
print("\nIevade (|01>+|11>)/sqrt2 amplitūdas:", x_super)
print("QFT(rezultāts) amplitūdas:", np.round(y_super, 3))

```
